```
##Implementing a simple AND gate using perceptron
import numpy as np
#input features
input=np.array([[0,0],[0,1],[1,0],[1,1]])
input.shape
```

```
print(input)
```

```
#output
target=np.array([[0],[0],[0],[1]])
print(target)
```

```
#weights
weights=np.array([[0.1],[0.2 ]])
print(weights)
```

```
#Bias and learning rate
bias=1
lr=0.05
```

```
def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```
#define the main logic of each iteration and in total there are 10000 iterations
for epoch in range(10000):
  inp_o=np.dot(input,weights) + bias
  out_o=sigmoid(inp_o)
  error=out_o - target
  derr_out= error
  derr_o_i=sigmoid_der(out_o)
  mul_der=derr_out * derr_o_i
  input_t=input.T
  final_der=np.dot(input_t,mul_der)
  weights-=lr*final_der
  for i in mul_der:
    bias-=lr*i
  print(weights)
  print(bias)
```

```
print(weights)
print(bias)
```

```
#Predicting output for 1,0

input=np.array([1,0])
result1=np.dot(input,weights) + bias
result2=sigmoid(result1)
```

```
print(result2)

```

Note: We can see that the predicted output is very close to 0.
